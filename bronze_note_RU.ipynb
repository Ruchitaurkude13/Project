```python
from pyspark.sql import SparkSession

# Spark session initialize 
spark = SparkSession.builder.appName("HR_Staging_L").getOrCreate()

# file paths
source_file_path = injection_lh_fpath+"//"+"HR_DATA.csv"  # Source CSV file
staging_file_path = injection_lh_fpath+"//"+"Hr_Data_Staging.csv"  # Staging layer file

# Load the raw HR data into a DataFrame
hr_stag_df = spark.read.option("header", "true").option("inferSchema","true").csv(source_file_path)

# Save the raw data into the Staging Layer as CSV
hr_stag_df.write.mode("overwrite").option("header", "true").csv(staging_file_path)

# hr_staging_df.show(5)

display(hr_stag_df)
```

```python
hr_stag_df.printSchema()
print("Number of rows ",hr_stag_df.count())
```

```python
from pyspark.sql.functions import col, sum
display(hr_stag_df.select([sum(col(c).isNull().cast("int")).alias(c) for c in hr_stag_df.columns])) 
```
